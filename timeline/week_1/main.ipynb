```python
import re
```

```python
INPUT="Nepal's Home Minister Ramesh Lekhak resigned Monday following the violence, Communications Minister Prithvi Subba confirmed to reporters late Monday."

OUTPUT=['Nepal', '’', 's', 'Home', 'Minister', 'Ramesh', 'Lekhak', 'resigned', 'Monday', 'following', 'the', 'violence', ',', 'Communications', 'Minister', 'Prithvi', 'Subba', 'confirmed', 'to', 'reporters', 'late', 'Monday', '.']

def split_text(text):
    pattern = r"([ .',])"
    text = re.split(pattern, text)
    text = [i for i in text if i != '' and i != ' ']
    return text

print(split_text(INPUT))
```

```python
INPUT="Nepal's Home Minister Ramesh Lekhak resigned Monday following the violence, Communications Minister Prithvi Subba confirmed to reporters late Monday."

OUTPUT=['Nepal', 'Home Minister Ramesh Lekkak', 'Monday', 'Communications Minister Prithvi Subba', 'Monday']

def get_identity(text):
    pattern = r"([A-Z][a-z]+(?:\s[A-Z][a-z]+)*)"
    text = re.findall(pattern, text)
    return text


get_identity(INPUT)
```

```python
INPUT="Dự kiến, số thu ngân sách từ thuế thu nhập cá nhân còn khoảng 84.477 tỷ đồng, giảm 21.000 tỷ mỗi năm so với hiện hành. Số người còn phải nộp thuế là 2,21 triệu người, giảm 2,18 triệu người (49,66%) do chuyển từ bậc 1 sang diện không chịu thuế."

OUTPUT=['Dự', '84.477', '21.000',  'Số', "2,21", "2,18",  '49,66']

def get_digit(text):
	pattern = r"[0-9]+(?:[,.][0-9]+)*|[A-Z]\w+"
	text = re.findall(pattern, text, flags=re.UNICODE)
	return text

get_digit(INPUT)
```
