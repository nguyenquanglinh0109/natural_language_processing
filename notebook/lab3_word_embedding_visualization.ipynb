```python
from gensim.models import Word2Vec
# import nltk
# from nltk.tokenize import word_tokenize
import re
import gensim.downloader
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
import plotly.express as px
import numpy as np
```

```python
word2vec_vector = gensim.downloader.load("word2vec-google-news-300")
glove_vector = gensim.downloader.load("glove-wiki-gigaword-300")
fasttext_vector = gensim.downloader.load("fasttext-wiki-news-subwords-300")
```

```python
def get_visual_vector(plot_words, vector):
  word_matrix = np.array([vector[word] for word in plot_words])
  num_components = 3
  pca = PCA(n_components=num_components)
  pca_result = pca.fit_transform(word_matrix)
  return pca_result

def visual_pca(vector, plot_words):
  data = {
    'x': vector[:, 0],
    'y': vector[:, 1],
    'z': vector[:, 2],
    'word': plot_words
  }

  fig = px.scatter_3d(data, x='x', y='y', z='z', text='word')

  fig.update_traces(marker=dict(size=8))

  fig.show()
```

# Word2Vec

```python
plot_words = ["king", "queen", "man", "woman", "glass","boy","girl"]

vector = get_visual_vector(plot_words, word2vec_vector)
visual_pca(vector, plot_words)
```

==> Quan sát biểu đồ trực quan có thể thấy được hướng của các cặp vector: man-woman, king-queen, girl-boy dường như rất gần nhau. Cho thấy khả năng biểu diễn vector ngữ nghĩa tương đối tốt của Word2Vec
